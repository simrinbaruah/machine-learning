```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier, GradientBoostingClassifier
from sklearn.metrics import accuracy_score
```

```python
df['Survived'].value_counts()
```

```python
df = df.set_index('PassengerId')
```

```python
df = df.drop(columns=['Name'])
df = df.drop(columns=['Ticket'])
```

```python
df = pd.get_dummies(df, columns=['Sex', 'Embarked'], dtype='int')
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
X = df.drop('Survived', axis=1)
y = df['Survived']
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, test_size=0.2)
```

```python
scaler = MinMaxScaler(feature_range=(0, 1))  # Normalize to range 0-1 (common choice)

scaler.fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
model1 = RandomForestClassifier(random_state=42)
model2 = AdaBoostClassifier(random_state=42)
model3 = GradientBoostingClassifier(random_state=42)
```

```python
model1.fit(X_train_scaled, y_train)
model2.fit(X_train_scaled, y_train)
model3.fit(X_train_scaled, y_train)
```

```python
predictions1 = model1.predict(X_test_scaled)
predictions2 = model2.predict(X_test_scaled)
predictions3 = model3.predict(X_test_scaled)
```

```python
trainpredictions1 = model1.predict(X_train_scaled)
trainpredictions2 = model2.predict(X_train_scaled)
trainpredictions3 = model3.predict(X_train_scaled)
```

```python
accuracy1 = accuracy_score(y_test, predictions1)
accuracy2 = accuracy_score(y_test, predictions2)
accuracy3 = accuracy_score(y_test, predictions3)

print("Accuracy of Model 1:", accuracy_score(y_train, trainpredictions1))
print("Accuracy of Model 2:", accuracy_score(y_train, trainpredictions2))
print("Accuracy of Model 3:", accuracy_score(y_train, trainpredictions3))

print("Accuracy of Model 1:", accuracy1)
print("Accuracy of Model 2:", accuracy2)
print("Accuracy of Model 3:", accuracy3)
```

```python
#Using AdaBoostClassifier

#Hyp tuning
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {
    'n_estimators': [50, 100, 200],  # Number of boosting stages (decision trees)
    'learning_rate': [0.1, 0.5, 1.0],  # Controls step size in each boosting iteration
    'max_depth': [3, 5, 8],  # Maximum depth of individual decision trees
    'subsample': [0.5, 0.8, 1.0]  # Fraction of samples used for training each tree
}
```

```python
model = GradientBoostingClassifier(random_state=42)
```

```python
grid_search = GridSearchCV(model, param_grid, scoring='accuracy', cv=5, verbose=2)
```

```python
grid_search.fit(X_train_scaled, y_train)
```

```python
grid_search.best_params_
```

```python
best_model = grid_search.best_estimator_
best_params = grid_search.best_params_
```

```python
predictions = best_model.predict(X_test_scaled)
```

```python
accuracy_score(predictions, y_test)
```
